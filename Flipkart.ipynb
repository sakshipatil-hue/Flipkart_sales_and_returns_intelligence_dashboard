```python
# Clean retail_price
#  Raw format: "₹2,999" or "2999.0" or NaN
# ─────────────────────────────────────────────────────────────
def clean_price(val):
    if pd.isnull(val):
        return np.nan
    val = str(val)
    val = re.sub(r'[₹,\s]', '', val)   # remove ₹ symbol, commas, spaces
    val = re.sub(r'[^\d.]', '', val)    # keep only digits and decimal point
    try:
        return float(val)
    except:
        return np.nan
 
df['retail_price'] = df['retail_price'].apply(clean_price)
df['discounted_price'] = df['discounted_price'].apply(clean_price)
 
```

```python

print("retail_price sample:")
print(df['retail_price'].describe())
print("\ndiscounted_price sample:")
print(df['discounted_price'].describe())
```

```python
#  CELL 7 — Extract main_category from product_category_tree
#  Raw format: "Electronics >> Mobiles >> Samsung Mobiles"
# ─────────────────────────────────────────────────────────────
def extract_main_category(val):
    if pd.isnull(val):
        return 'Unknown'
    val = str(val)
    # Remove brackets and quotes that Kaggle dataset wraps around this field
    val = val.strip('[]"\'')
    # Split on >> and take the first part
    parts = val.split('>>')
    return parts[0].strip() if parts else 'Unknown'
 
def extract_sub_category(val):
    if pd.isnull(val):
        return 'Unknown'
    val = str(val).strip('[]"\'')
    parts = val.split('>>')
    return parts[1].strip() if len(parts) > 1 else 'Unknown'
 
df['main_category'] = df['product_category_tree'].apply(extract_main_category)
df['sub_category']  = df['product_category_tree'].apply(extract_sub_category)
 
print("Top 15 main categories:")
print(df['main_category'].value_counts().head(15))
```

```python
#  CELL 8 — Clean brand column
# ─────────────────────────────────────────────────────────────
df['brand'] = df['brand'].astype(str).str.strip().str.title()
df['brand'] = df['brand'].replace({'Nan': 'Unknown', 'None': 'Unknown', '': 'Unknown'})
 
print("Top 20 brands:")
print(df['brand'].value_counts().head(20))
```

```python
#  CELL 9 — Drop rows with critical nulls
#  We need at least a price and category to do analysis
# ─────────────────────────────────────────────────────────────
print(f"Before drop: {df.shape[0]} rows")
 
df = df.dropna(subset=['retail_price', 'discounted_price'])
df = df[df['retail_price'] > 0]
df = df[df['discounted_price'] > 0]
 
# Remove impossible prices (discounted > retail means bad data)
df = df[df['discounted_price'] <= df['retail_price']]
 
print(f"After drop:  {df.shape[0]} rows")
print(f"Removed:     {df.shape[0]} rows with bad price data")
```

```python
df['product_id'] = ['P' + str(i+1).zfill(5) for i in range(len(df))]
# P00001, P00002, P00003 ... clean and readable
```

```python
#  STEP 2 — Generate dim_customers (10,000 Indian customers)
# ─────────────────────────────────────────────────────────────
 
# Real Indian cities with realistic population weighting
CITIES = {
    'Mumbai':       ('Maharashtra', 0.12),
    'Delhi':        ('Delhi',       0.11),
    'Bengaluru':    ('Karnataka',   0.10),
    'Hyderabad':    ('Telangana',   0.08),
    'Chennai':      ('Tamil Nadu',  0.07),
    'Kolkata':      ('West Bengal', 0.07),
    'Pune':         ('Maharashtra', 0.06),
    'Ahmedabad':    ('Gujarat',     0.05),
    'Jaipur':       ('Rajasthan',   0.04),
    'Lucknow':      ('UP',          0.04),
    'Bhopal':       ('MP',          0.03),
    'Indore':       ('MP',          0.03),
    'Surat':        ('Gujarat',     0.03),
    'Nagpur':       ('Maharashtra', 0.03),
    'Patna':        ('Bihar',       0.02),
    'Coimbatore':   ('Tamil Nadu',  0.02),
    'Bhubaneswar':  ('Odisha',      0.02),
    'Kochi':        ('Kerala',      0.02),
    'Chandigarh':   ('Punjab',      0.02),
    'Guwahati':     ('Assam',       0.02),
}
 
city_names   = list(CITIES.keys())
city_weights = [CITIES[c][1] for c in city_names]
 
# Indian first + last names for realism
FIRST_NAMES = [
    'Aarav','Vivaan','Aditya','Vihaan','Arjun','Sai','Reyansh',
    'Ayaan','Krishna','Ishaan','Ananya','Diya','Priya','Sneha',
    'Pooja','Neha','Kavya','Riya','Meera','Sakshi','Rahul',
    'Rohit','Amit','Vikram','Suresh','Ramesh','Deepak','Raj',
    'Nikhil','Kunal','Shruti','Divya','Swati','Rekha','Sunita'
]
LAST_NAMES = [
    'Sharma','Verma','Singh','Patel','Kumar','Gupta','Joshi',
    'Mishra','Yadav','Tiwari','Nair','Iyer','Reddy','Rao',
    'Mehta','Shah','Jain','Agarwal','Bose','Chatterjee',
    'Pillai','Menon','Desai','Patil','Shukla','Pandey'
]
 
N_CUSTOMERS = 10000
customer_ids = ['C' + str(i+1).zfill(6) for i in range(N_CUSTOMERS)]
cities_chosen = random.choices(city_names, weights=city_weights, k=N_CUSTOMERS)
 
dim_customers = pd.DataFrame({
    'customer_id':   customer_ids,
    'customer_name': [
        random.choice(FIRST_NAMES) + ' ' + random.choice(LAST_NAMES)
        for _ in range(N_CUSTOMERS)
    ],
    'city':  cities_chosen,
    'state': [CITIES[c][0] for c in cities_chosen],
    'age':   np.random.choice(
        range(18, 65),
        size=N_CUSTOMERS,
        p=np.array([
            0.03 if a < 22 else
            0.04 if a < 30 else
            0.035 if a < 40 else
            0.02 if a < 50 else
            0.01
            for a in range(18, 65)
        ]) / sum([
            0.03 if a < 22 else
            0.04 if a < 30 else
            0.035 if a < 40 else
            0.02 if a < 50 else
            0.01
            for a in range(18, 65)
        ])
    ),
    'customer_segment': np.random.choice(
        ['Budget Shopper','Regular','Premium','Loyal'],
        size=N_CUSTOMERS,
        p=[0.35, 0.40, 0.15, 0.10]
    )
})
 
print(f"Customers generated: {len(dim_customers):,}")
```

```python
#  STEP 3 — Generate dim_sellers (500 sellers)
# ─────────────────────────────────────────────────────────────
 
SELLER_NAMES = [
    'TechMart India','QuickShop','RetailKing','BestBuy India',
    'MegaStore','ShopEasy','ValueMart','PrimeSeller','TopDeals',
    'FlashSale India','TrustSeller','ReliableMart','IndiaShop',
    'DealZone','SmartBuy','BargainHub','QualityFirst','FastShip',
    'MumbaiMart','DelhiDeals','BengaluruBuy','HydStore','ChennaiShop'
]
 
N_SELLERS = 500
dim_sellers = pd.DataFrame({
    'seller_id':     ['S' + str(i+1).zfill(4) for i in range(N_SELLERS)],
    'seller_name':   [
        random.choice(SELLER_NAMES) + ' ' + str(random.randint(1,999))
        for _ in range(N_SELLERS)
    ],
    'seller_city':   random.choices(city_names, weights=city_weights, k=N_SELLERS),
    'seller_rating': np.round(np.random.normal(4.1, 0.5, N_SELLERS).clip(1, 5), 1),
    'joined_year':   np.random.choice(range(2015, 2024), size=N_SELLERS,
                        p=[0.05,0.08,0.10,0.12,0.15,0.15,0.15,0.12,0.08])
})
 
print(f"Sellers generated:   {len(dim_sellers):,}")
```

```python
#  STEP 4 — Generate fact_orders (100,000 orders over 2 years)
#
#  Key realism features:
#  - Big Billion Days spike (Oct): 3x normal volume
#  - Diwali spike (Oct-Nov): 2x normal
#  - Weekend boost: 1.3x
#  - Category-based pricing logic
#  - Customer segment affects purchase behavior
# ─────────────────────────────────────────────────────────────
 
START_DATE = datetime(2022, 1, 1)
END_DATE   = datetime(2023, 12, 31)
N_ORDERS   = 100000
 
# Date generation with festive season weighting
def generate_order_dates(n):
    dates = []
    current = START_DATE
    date_pool = []
    weights_pool = []
 
    while current <= END_DATE:
        month = current.month
        # Base weight
        weight = 1.0
        # Festive season boost
        if month == 10:  weight = 3.5   # Big Billion Days
        elif month == 11: weight = 2.5  # Post-Diwali
        elif month == 12: weight = 1.8  # Christmas/New Year
        elif month == 1:  weight = 1.5  # Republic Day sale
        elif month in [4,5]: weight = 1.3  # Summer sale
        # Weekend boost
        if current.weekday() >= 5: weight *= 1.3
        date_pool.append(current)
        weights_pool.append(weight)
        current += timedelta(days=1)
 
    chosen = random.choices(date_pool, weights=weights_pool, k=n)
    return chosen
 
print("Generating order dates with festive season patterns...")
order_dates = generate_order_dates(N_ORDERS)
 
# Sample products weighted by category popularity
category_weights_map = {
    'Clothing':         0.20,
    'Electronics':      0.18,
    'Footwear':         0.10,
    'Furniture':        0.08,
    'Jewellery':        0.07,
    'Home Furnishing':  0.07,
    'Kitchen':          0.06,
    'Sports':           0.05,
    'Toys & School':    0.04,
    'Beauty':           0.04,
    'Mobiles':          0.04,
    'Other':            0.07,
}
 
# Assign product weights based on category
dim_products['pop_weight'] = dim_products['main_category'].map(
    lambda c: next(
        (v for k,v in category_weights_map.items() if k.lower() in c.lower()),
        0.03
    )
)
product_probs = dim_products['pop_weight'] / dim_products['pop_weight'].sum()
 
print("Sampling products for orders...")
sampled_products = dim_products.sample(
    n=N_ORDERS, replace=True, weights=product_probs
).reset_index(drop=True)
 
# Generate quantities (most orders = 1 item)
quantities = np.random.choice([1,2,3,4,5], size=N_ORDERS, p=[0.70,0.17,0.07,0.04,0.02])
 
# Payment methods — UPI dominates in India
payment_methods = np.random.choice(
    ['UPI','Credit Card','Debit Card','Net Banking','COD','EMI'],
    size=N_ORDERS,
    p=[0.38, 0.18, 0.15, 0.10, 0.12, 0.07]
)
 
# Order status — most delivered, some cancelled/returned
order_statuses = np.random.choice(
    ['Delivered','Delivered','Delivered','Delivered',
     'Cancelled','Returned','Processing'],
    size=N_ORDERS
)
 
# Build fact_orders
fact_orders = pd.DataFrame({
    'order_id':       ['ORD' + str(i+1).zfill(7) for i in range(N_ORDERS)],
    'order_date':     [d.strftime('%Y-%m-%d') for d in order_dates],
    'order_month':    [d.month for d in order_dates],
    'order_year':     [d.year for d in order_dates],
    'order_quarter':  [f"Q{(d.month-1)//3+1}" for d in order_dates],
    'is_weekend':     [1 if d.weekday() >= 5 else 0 for d in order_dates],
 
    # Big Billion Days = Oct 1-7
    'is_big_billion_day': [
        1 if (d.month == 10 and d.day <= 7) else 0 for d in order_dates
    ],
    # Festive season = Oct + Nov
    'is_festive_season': [
        1 if d.month in [10, 11] else 0 for d in order_dates
    ],
 
    'customer_id':    random.choices(dim_customers['customer_id'].tolist(), k=N_ORDERS),
    'product_id':     sampled_products['product_id'].tolist(),
    'seller_id':      random.choices(dim_sellers['seller_id'].tolist(), k=N_ORDERS),
 
    'quantity':       quantities,
    'unit_price':     sampled_products['discounted_price'].round(2).tolist(),
    'retail_price':   sampled_products['retail_price'].round(2).tolist(),
    'total_amount':   (sampled_products['discounted_price'] * quantities).round(2).tolist(),
    'discount_pct':   sampled_products['discount_pct'].round(2).tolist(),
 
    'payment_method': payment_methods,
    'order_status':   order_statuses,
 
    'main_category':  sampled_products['main_category'].tolist(),
})
 
print(f"Orders generated:    {len(fact_orders):,}")
print(f"  Big Billion Days orders: {fact_orders['is_big_billion_day'].sum():,}")
print(f"  Festive season orders:   {fact_orders['is_festive_season'].sum():,}")
print(f"  Returned orders:         {(fact_orders['order_status']=='Returned').sum():,}")
 
```

```python
#  STEP 5 — Generate fact_returns
#  Only for orders with status = 'Returned'
# ─────────────────────────────────────────────────────────────
 
returned_orders = fact_orders[fact_orders['order_status'] == 'Returned'].copy()
 
RETURN_REASONS = {
    'Clothing':    ['Size issue','Wrong item delivered','Quality not as expected','Changed mind'],
    'Electronics': ['Defective product','Not as described','Better price elsewhere','Damaged in transit'],
    'Footwear':    ['Size issue','Quality issue','Wrong item','Damaged'],
    'Furniture':   ['Damaged in transit','Wrong item','Assembly issue','Quality issue'],
    'default':     ['Quality not as expected','Wrong item','Changed mind',
                    'Defective product','Better price elsewhere','Damaged in transit']
}
 
def get_return_reason(category):
    reasons = RETURN_REASONS.get(category, RETURN_REASONS['default'])
    return random.choice(reasons)
 
return_dates = [
    datetime.strptime(d, '%Y-%m-%d') + timedelta(days=random.randint(1, 10))
    for d in returned_orders['order_date']
]
 
fact_returns = pd.DataFrame({
    'return_id':     ['RET' + str(i+1).zfill(6) for i in range(len(returned_orders))],
    'order_id':      returned_orders['order_id'].values,
    'product_id':    returned_orders['product_id'].values,
    'customer_id':   returned_orders['customer_id'].values,
    'main_category': returned_orders['main_category'].values,
    'return_date':   [d.strftime('%Y-%m-%d') for d in return_dates],
    'return_reason': [
        get_return_reason(cat) for cat in returned_orders['main_category'].values
    ],
    'refund_amount': returned_orders['total_amount'].values,
    'return_status': np.random.choice(
        ['Refunded','Replacement Sent','Processing'],
        size=len(returned_orders),
        p=[0.65, 0.25, 0.10]
    )
})
 
print(f"Returns generated:   {len(fact_returns):,}")
```
